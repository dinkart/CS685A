```python
import pandas as pd
```

Only Men's Calculation

```python
#Only Men Count
cols = ["state-code", "area-name", "Total", "Age-group", "Tot_p", "Illiterate", "Literate but below primary", "Primary but below middle", "Middle but below matric/secondary", "Matric", "HigherSecondary", "NonTechDiploma", "TechDiploma", "Graduate and above"]
df1 = pd.read_excel("../Data/DDW-0000C-08.xlsx", skiprows = [0,1,2,3,5,6], names = cols, usecols = [1, 3, 4, 5, 7, 10, 19, 22, 25, 28, 31, 34, 37, 40])
df1 = df1[(df1["Total"] == "Total") & (df1["Age-group"] == "All ages")]
```

```python
# Adding "Matric/Secondary but below graduate" column to C-08 data
temp = []
for idx, row in df1.iterrows():
    temp.append(row["Matric"] + row["HigherSecondary"] + row["NonTechDiploma"] + row["TechDiploma"])
df1["Matric/Secondary but below graduate"] = temp
```

```python
cols = ["state-code", "area-name", "Total", "EduLevel", "TwoLang_m", "TwoLang_f"]
df2 = pd.read_excel("../Data/DDW-C19-0000.xlsx", skiprows = [0,1,2,3,5], usecols = [0, 2, 3, 4, 6, 7], names = cols)
df2 = df2[df2['Total'] == "Total"]
df2 = df2[(df2["EduLevel"] != "Total") & (df2["EduLevel"] != "Literate")]
df2["state-code"] = df2[["state-code"]].apply(pd.to_numeric)
```

```python
output = {"state/ut": [], "literacy-group-males": [], "ratio-males": [], "literacy-group-females": [], "ratio-females": []}
for idx, row in df1.iterrows():
    temp = df2[df2["state-code"] == row["state-code"]]
    output["state/ut"].append(temp["state-code"].unique()[0])
    litgrp_m = ""
    maxpercent_m = 0
    for idx2, row2 in temp.iterrows():
        threeLang = row[row2["EduLevel"]] - row2["TwoLang_m"]
        tot_p = row[row2["EduLevel"]]
        percent = threeLang/tot_p
        if percent > maxpercent_m:
            maxpercent_m = percent
            litgrp_m = row2["EduLevel"]
    output["literacy-group-males"].append(litgrp_m)
    output["ratio-males"].append(maxpercent_m)
```

Only Female's Count

```python
#Only Female Count
cols = ["state-code", "area-name", "Total", "Age-group", "Tot_p", "Illiterate", "Literate but below primary", "Primary but below middle", "Middle but below matric/secondary", "Matric", "HigherSecondary", "NonTechDiploma", "TechDiploma", "Graduate and above"]
df1 = pd.read_excel("../Data/DDW-0000C-08.xlsx", skiprows = [0,1,2,3,5,6], names = cols, usecols = [1, 3, 4, 5, 8, 11, 20, 23, 26, 29, 32, 35, 38, 41])
df1 = df1[(df1["Total"] == "Total") & (df1["Age-group"] == "All ages")]
```

```python
# Adding "Matric/Secondary but below graduate" column to C-08 data
temp = []
for idx, row in df1.iterrows():
    temp.append(row["Matric"] + row["HigherSecondary"] + row["NonTechDiploma"] + row["TechDiploma"])
df1["Matric/Secondary but below graduate"] = temp
```

```python
for idx, row in df1.iterrows():
    temp = df2[df2["state-code"] == row["state-code"]]
    litgrp_f = ""
    maxpercent_f = 0
    for idx2, row2 in temp.iterrows():
        threeLang = row[row2["EduLevel"]] - row2["TwoLang_f"]
        tot_p = row[row2["EduLevel"]]
        percent = threeLang/tot_p
        if percent > maxpercent_f:
            maxpercent_f = percent
            litgrp_f = row2["EduLevel"]
    output["literacy-group-females"].append(litgrp_f)
    output["ratio-females"].append(maxpercent_f)
```

```python
df = pd.DataFrame(output).sort_values(by=['state/ut'])
df.to_csv("../Final Outputs/literacy-gender-c.csv", index = False)
```
