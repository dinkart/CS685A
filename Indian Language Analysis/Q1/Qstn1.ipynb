```python
import pandas as pd
```

```python
df1 = pd.read_excel("../Data/DDW_PCA0000_2011_Indiastatedist.xlsx")
df1 = df1[(df1["TRU"] == "Total") & (df1["Level"] == "STATE")]
df2 = pd.read_excel("../Data/DDW-C18-0000.xlsx", skiprows = [0,1,2,3,5])
df2 = df2[(df2["4"] == "Total") & (df2["5"] == "Total")]
```

```python
output = {"state-code" : [], "percent-one" : [], "percent-two" : [], "percent-three" : []}
for idx,row in df1.iterrows():
    temp = df2[df2['1'] == row["State"]]
    output["state-code"].append(row["State"])
    output["percent-one"].append(float(row["TOT_P"] - temp["6"])/row["TOT_P"] * 100)
    output["percent-two"].append(float(temp["6"] - temp["9"])/row["TOT_P"] * 100)
    output["percent-three"].append(float(temp["9"])/row["TOT_P"] * 100)
```

```python
df = pd.DataFrame(output).sort_values(by=['state-code'])
df.to_csv("../Final Outputs/percent-india.csv", index = False)
```
