```python
import pandas as pd
```

```python
cols = ["state-code", "area-name", "age-groups", "persons", "males", "females"]
df1 = pd.read_excel("../Data/DDW-0000C-14.xls", skiprows = [0,1,2,3,5,6], names = cols, usecols = [1, 3, 4, 5, 6, 7])
df1.replace(["All ages"], ["Total"], inplace = True)
```

```python
#Adding age group "30-49"
temp = df1[(df1["age-groups"] == "30-34") | (df1["age-groups"] == "35-39") | (df1["age-groups"] == "40-44") | (df1["age-groups"] == "45-49")]
temp = temp.groupby(["state-code", "area-name"], as_index=False).agg('sum')
agegrp = ["30-49" for i in range(len(temp))]
temp["age-groups"] = agegrp
df1 = df1.append(temp, ignore_index = True)
```

```python
#Adding age group "50-69"
temp = df1[(df1["age-groups"] == "50-54") | (df1["age-groups"] == "55-59") | (df1["age-groups"] == "60-64") | (df1["age-groups"] == "65-69")]
temp = temp.groupby(["state-code", "area-name"], as_index=False).agg('sum')
agegrp = ["50-69" for i in range(len(temp))]
temp["age-groups"] = agegrp
df1 = df1.append(temp, ignore_index = True)
```

```python
#Adding age group "70+"
temp = df1[(df1["age-groups"] == "70-74") | (df1["age-groups"] == "75-79") | (df1["age-groups"] == "80+")]
temp = temp.groupby(["state-code", "area-name"], as_index=False).agg('sum')
agegrp = ["70+" for i in range(len(temp))]
temp["age-groups"] = agegrp
df1 = df1.append(temp, ignore_index = True)
```

```python
cols = ["state-code", "area-name", "Total", "age-groups", "persons_two", "males_two", "females_two", "persons_three", "males_three", "females_three"]
df2 = pd.read_excel("../Data/DDW-C18-0000.xlsx", skiprows = [0,1,2,3,5], usecols = [0, 2, 3, 4, 5, 6, 7, 8, 9, 10], names = cols)
df2 = df2[df2['Total'] == "Total"]
df2 = df2[df2["age-groups"] != "Age not stated"]
df2.reset_index(inplace = True)
#Getting exact two language speaking males and females count
df2["person_exact_two"] = [df2["persons_two"][i]-df2["persons_three"][i] for i in range(df2.shape[0])]
df2["males_exact_two"] = [df2["males_two"][i]-df2["males_three"][i] for i in range(df2.shape[0])]
df2["females_exact_two"] = [df2["females_two"][i]-df2["females_three"][i] for i in range(df2.shape[0])]
```

```python
output = {"state/ut": [], "age-group-males": [], "ratio-males": [], "age-group-females": [], "ratio-females": []}
scodes = df2["state-code"].unique().tolist()
for code in scodes:
    temp = df2[df2["state-code"] == code]
    output["state/ut"].append(temp["state-code"].unique()[0])
    agegrp_m = ""
    maxpercent_m = 0
    agegrp_f = ""
    maxpercent_f = 0
    for idx, row in temp.iterrows():
        tot_m = df1["males"][(df1["state-code"] == code) & (df1["age-groups"] == row['age-groups'])].tolist()[0]
        percent = row["males_exact_two"]/tot_m
        if percent > maxpercent_m:
            maxpercent_m = percent
            agegrp_m = row["age-groups"]
        
        tot_f = df1["females"][(df1["state-code"] == code) & (df1["age-groups"] == row['age-groups'])].tolist()[0]
        percent = row["females_exact_two"]/tot_f
        if percent > maxpercent_f:
            maxpercent_f = percent
            agegrp_f = row["age-groups"]
    output["age-group-males"].append(agegrp_m)
    output["ratio-males"].append(maxpercent_m)
    output["age-group-females"].append(agegrp_f)
    output["ratio-females"].append(maxpercent_f)
```

```python
df = pd.DataFrame(output).sort_values(by=['state/ut'])
df.to_csv("../Final Outputs/age-gender-b.csv", index = False)
```
