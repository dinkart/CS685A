```python
import pandas as pd
```

```python
cols = ["state-code", "area-name", "age-groups", "persons"]
df1 = pd.read_excel("../Data/DDW-0000C-14.xls", skiprows = [0,1,2,3,5,6], names = cols, usecols = [1, 3, 4, 5])
df1.replace(["All ages"], ["Total"], inplace = True)
```

```python
#Adding age group "30-49"
temp = df1[(df1["age-groups"] == "30-34") | (df1["age-groups"] == "35-39") | (df1["age-groups"] == "40-44") | (df1["age-groups"] == "45-49")]
temp = temp.groupby(["state-code", "area-name"], as_index=False).agg('sum')
agegrp = ["30-49" for i in range(len(temp))]
temp["age-groups"] = agegrp
df1 = df1.append(temp, ignore_index = True)
```

```python
#Adding age group "50-69"
temp = df1[(df1["age-groups"] == "50-54") | (df1["age-groups"] == "55-59") | (df1["age-groups"] == "60-64") | (df1["age-groups"] == "65-69")]
temp = temp.groupby(["state-code", "area-name"], as_index=False).agg('sum')
agegrp = ["50-69" for i in range(len(temp))]
temp["age-groups"] = agegrp
df1 = df1.append(temp, ignore_index = True)
```

```python
#Adding age group "70+"
temp = df1[(df1["age-groups"] == "70-74") | (df1["age-groups"] == "75-79") | (df1["age-groups"] == "80+")]
temp = temp.groupby(["state-code", "area-name"], as_index=False).agg('sum')
agegrp = ["70+" for i in range(len(temp))]
temp["age-groups"] = agegrp
df1 = df1.append(temp, ignore_index = True)
```

```python
df2 = pd.read_excel("../Data/DDW-C18-0000.xlsx", skiprows = [0,1,2,3,5])
df2 = df2[df2['4'] == "Total"]
```

```python
output = {"state/ut": [], "age-group": [], "percentage": []}
scodes = df2["1"].unique().tolist()
for code in scodes:
    temp = df2[df2["1"] == code]
    output["state/ut"].append(temp["1"].unique()[0])
    agegrp = ""
    maxpercent = 0
    for idx, row in temp.iterrows():
        tot_p = df1["persons"][(df1["state-code"] == code) & (df1["age-groups"] == row['5'])].tolist()[0]
        percent = row["9"]/tot_p
        if percent > maxpercent:
            maxpercent = percent
            agegrp = row["5"]
    output["age-group"].append(agegrp)
    output["percentage"].append(maxpercent*100)
```

```python
df = pd.DataFrame(output).sort_values(by=['state/ut'])
df.to_csv("../Final Outputs/age-india.csv", index = False)
```
